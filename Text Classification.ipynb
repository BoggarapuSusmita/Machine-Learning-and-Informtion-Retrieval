```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
def cleaning_data(review):
    lemmatizer = WordNetLemmatizer()
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = nltk.word_tokenize(review)
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = [word for word in review if not word == 'br']
    review = ' '.join(review)
    return review
```

```python
for i in range(len(df)):
    df['review'][i]=cleaning_data(df['review'][i])

print(df)
```

```python
from sklearn.model_selection import train_test_split
train_val, test, = train_test_split(df, test_size = 0.20, random_state = 0)
train, validation, = train_test_split(train_val, test_size = 0.25, random_state = 0)
print(len(train),len(test),len(validation))
```

```python
pos_review = pos['review'].tolist()
neg_review = neg['review'].tolist()
print(len(pos_review), len(neg_review))
```

```python
def get_adjectives(review):
    adj_list=[]
    for i in review:
        words = nltk.word_tokenize(i)
        words = nltk.pos_tag(words)
    
        for w in words:
            if w[1][0] == 'J':
                adj_list.append(w[0])
    
    return adj_list
```

```python
adj_words_pos=get_adjectives(pos_review)
adj_words_neg=get_adjectives(neg_review)

adj_words_neg
```

```python
def data_adj(i):
    adj_words = []
    words = nltk.word_tokenize(i)
    words = nltk.pos_tag(words)
    
    for w in words:
        if w[1][0] == 'J':
            adj_words.append(w[0])
    
    return adj_words
```

```python
test_review = test['review'].tolist()
test_sentiment = test['sentiment'].tolist()
for i in range(len(test_review)):
    
    test_review[i] = data_adj(test_review[i])
    print(test_sentiment[i])
    print(test_review[i])
```

```python
adj_words_pos = nltk.FreqDist(adj_words_pos)
adj_words_neg = nltk.FreqDist(adj_words_neg)
print(len(adj_words_neg), len(adj_words_pos))

vocab_pos = len(adj_words_pos)
vocab_neg = len(adj_words_neg)

pos_n = sum(adj_words_pos.values())
neg_n = sum(adj_words_neg.values())

adj_words_pos
```

```python
#A priori probabilities
pos_porb = len(pos_review)/len(train)
neg_porb = len(neg_review)/len(train)
pos_porb, neg_porb
```

```python
#Class Conditional probabilities : (nk + 1 / n + vocab)
for i in adj_words_pos:
    adj_words_pos[i] = (adj_words_pos[i] + 1) / (pos_n + vocab_pos)

for i in adj_words_neg:
    adj_words_neg[i] = (adj_words_neg[i] + 1) / (neg_n + vocab_neg)

adj_words_neg
```

```python
adj_words_pos['how']
```

```python
def class_belongingness(review):
    class1, class2 = pos_porb, neg_porb
    
    for j in review:
        
        #positive class
        if adj_words_pos[j] == 0:
            class1 *= 1/(pos_n + vocab_pos)
        else:
            class1 *= adj_words_pos[j]
            
        #negative class
        if adj_words_neg[j] == 0:
            class2 *= 1/(neg_n + vocab_neg)
        else:
            class2 *= adj_words_neg[j]
            
    #class belongingness
    if class1 > class2:
        c = 1
    else :
        c = 0
    
    return c
```

```python
#Posterior probabilities on validation data

val_review = validation['review'].tolist()
val_sentiment = validation['sentiment'].tolist()

val_sentiment_pred=[]

for i in range(len(val_review)):
    
    val_review[i] = data_adj(val_review[i])
    
    val_sentiment_pred.append(class_belongingness(val_review[i]))
    if val_sentiment[i]=='positive':
        val_sentiment[i]=1
    else :
        val_sentiment[i]=0

```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, confusion_matrix
accuracy=accuracy_score(val_sentiment, val_sentiment_pred)

print('ACCURCY :',accuracy)
```

```python
#Posterior probabilities on test data
test_sentiment_pred=[]
for i in range(len(test_review)):
    
    test_sentiment_pred.append(class_belongingness(test_review[i]))
    if test_sentiment[i]=='positive':
        test_sentiment[i]=1
    else :
        test_sentiment[i]=0
```

```python
c_m=confusion_matrix(test_sentiment, test_sentiment_pred)
accuracy=accuracy_score(test_sentiment, test_sentiment_pred)
precision=precision_score(test_sentiment, test_sentiment_pred)
recall=recall_score(test_sentiment, test_sentiment_pred)

print(c_m)
print('ACCURCY :',accuracy)
print('ERROR RATE :',1-accuracy)
print('PRECISION :',precision)
print('RECALL :',recall)
```
