```python
#NORMALIZATION
def Normalization(column):
    column=column-min(column)
    column=column/(max(column)-min(column))
    return column

X=pd.DataFrame()
for i in num_cols:
    X[i]=Normalization(df[i].to_numpy(dtype='float'))
```

```python
test=X[77:78]
test=np.resize(test,test.size)
```

```python
test_class=test[-1]
print(test[-1])
test=np.resize(test,test.size-1)
```

```python
X=X.to_numpy()
```

```python
import math
dist=[]
for ind in X:
    sum1=0
    for i in range(len(test)):
        #print(ind[i],test[i],i)
        sum1+=(ind[i]-test[i])**2
    dist.append(math.sqrt(sum1))
dist       
```

```python
dist_dict={}
for i in range(len(dist)):
    dist_dict[i]=[dist[i],X[i][8]]
dist_dict
```

```python
K=10
dist.sort()
maximum=dist[K]
class1,class2=0,0
```

```python
for i in dist_dict.values():
    #print(i[0],i[1])
    if i[0]<maximum and i[1]==0.0:
        class1+=1
    elif i[0]<maximum and i[1]==1.0:
        class2+=1
print(class1,class2)
if class1>class2:
    print("No Diabeties")
else:
    print("Has Diabeties")
```
